```python
from sklearn import datasets

import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
from sklearn import datasets

# 以 iris 資料庫為例
iris = datasets.load_iris( )
X = iris.data
y = iris.target

print( iris )
```

Axes3D( fig, rect, azim, elev )

- rect : ( float, float, float, float )   
   The ( left, bottom, width, height ) axes position.
- azim : float, optional  
   Azimuthal viewing angle, defaults to -60. ( 3D 圖形的方位角度 ) 
- elev : float, optional  
   Elevation viewing angle, defaults to 30. ( 3D 圖形的仰角數 )

- Scatter( ) : https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.scatter.html

### 3D Scatter Plot

```python
# 建立畫布
fig = plt.figure( num = 1 )#, figsize = (4, 3) )

# 將 figure 轉為 3D 圖
ax = Axes3D( fig )

# 3D 散佈圖
ax.scatter( xs = X[ :, 3 ],                 #  x 軸座標 
                ys = X[ :, 0 ],                 #  y 軸座標
                zs = X[ :, 2 ],                 #  z 軸座標
                c = y.astype( np.float ),   # 資料點的顏色
                edgecolor = 'face' )        # 資料點的輪廓顏色


# 加入標籤說明
for name, label in [ ( 'setosa', 0 ), ( 'versicolor', 1 ), ( 'virginica', 2 ) ] :
    ax.text3D( x = X[ y == label, 3 ].mean( ),               # 標籤的 x 軸位置 
                   y = X[ y == label, 0 ].mean( ),               # 標籤的 y 軸位置
                   z = X[ y == label, 2 ].mean( ) + 2,         # 標籤的 z 軸位置
                   s = name,                                         # 標籤內容
                   horizontalalignment = 'center',
                   bbox = dict( alpha = .2, edgecolor = 'w', facecolor = 'w' ) )


# 設定座標軸的刻度
#ax.w_xaxis.set_ticklabels( [ ] )
ax.w_yaxis.set_ticklabels( [ ] )
#ax.w_zaxis.set_ticklabels( [ ] )

# 座標軸的標籤
ax.set_xlabel( 'X Label' )
ax.set_ylabel( 'Y Label' )
ax.set_zlabel( 'Z Label' )

# 3D 圖的標題
ax.set_title( 'IRIS Data' )
ax.dist = 12                    # 標題與3D圖的間距

plt.show( )
```
