```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

##### 隨機生成 ( -1, 1 ) 的隨機矩陣
https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.random.html

```python
"""
Your Code Here
"""
matrix = -1 + 2*np.random.random( ( 10, 10 ) )

plt.figure( figsize = ( 10, 10 ) )
"""
Your Code Here
"""
sns.heatmap( matrix, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = -1 + 2*np.random.random( ( nrow , ncol ) )

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice( [ 0, 1, 2 ], size = nrow )
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper( plt.scatter , alpha = 0.2)
grid.map_diag( plt.hist )
grid.map_lower( sns.kdeplot , cmap = plt.cm.OrRd_r )

plt.show()
```

##### 隨機生成符合常態分布的隨機矩陣
https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.randn.html

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.randn( nrow, ncol )

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper( plt.scatter, alpha = 0.2)
grid.map_diag( plt.hist )
grid.map_lower( sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
